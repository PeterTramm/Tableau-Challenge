```python
#Import modules
import os
import pandas as pd 
import sys
import numpy as np
import math

#Set print options
np.set_printoptions(threshold=sys.maxsize)
```

```python
#Load data 

path = "Resources/JC-202312-citibike-tripdata.csv"
path1 = "Resources/JC-202311-citibike-tripdata.csv"
path2 = "Resources/JC-202310-citibike-tripdata.csv" 

Bike_2023_12 = pd.read_csv(path)
Bike_2023_11 = pd.read_csv(path1)
Bike_2023_10 = pd.read_csv(path2)
```

```python
# Check Counts and columns
print("Count of rows of dataframe: " + str(len(Bike_2023_12["ride_id"].unique())))
print("Rideable types: " + str(len(Bike_2023_12["rideable_type"].unique())))
print("Station Name count: " + str(len(Bike_2023_12["start_station_name"].unique())))
print("Ride Type: " + Bike_2023_12["rideable_type"].unique())
print("Member type: " + Bike_2023_12["member_casual"].unique())
```

```python
#Change date time 
Bike_2023_12['started_at'] = pd.to_datetime(Bike_2023_12["started_at"])
Bike_2023_12['ended_at'] = pd.to_datetime(Bike_2023_12["ended_at"])
Bike_2023_11['started_at'] = pd.to_datetime(Bike_2023_11["started_at"])
Bike_2023_11['ended_at'] = pd.to_datetime(Bike_2023_11["ended_at"])
Bike_2023_10['started_at'] = pd.to_datetime(Bike_2023_10["started_at"])
Bike_2023_10['ended_at'] = pd.to_datetime(Bike_2023_10["ended_at"])
```

```python
#Checking for null values
Bike_2023_12.isnull().sum()
```

```python
#Drop nulls

drop_12 = Bike_2023_12.dropna()
drop_11 = Bike_2023_11.dropna()
drop_10 = Bike_2023_10.dropna()


print(len(drop_12))
print(len(drop_11))
print(len(drop_10))
```

```python
# Cut latitude and longitude down to 2 decimal places
drop_12["start_lat"] = np.floor(drop_12['start_lat'] * 1000) / 1000
drop_12["start_lng"] = np.floor(drop_12["start_lng"] * 1000) / 1000
drop_12["end_lat"] = np.floor(drop_12["end_lat"] * 1000) / 1000
drop_12["end_lng"] = np.floor(drop_12["end_lng"] * 1000) / 1000
```

```python
# Cut latitude and longitude down to 2 decimal places
drop_11["start_lat"] = np.floor(drop_11['start_lat'] * 1000) / 1000
drop_11["start_lng"] = np.floor(drop_11["start_lng"] * 1000) / 1000
drop_11["end_lat"] = np.floor(drop_11["end_lat"] * 1000) / 1000
drop_11["end_lng"] = np.floor(drop_11["end_lng"] * 1000) / 1000
```

```python
# Cut latitude and longitude down to 2 decimal places
drop_10["start_lat"] = np.floor(drop_10['start_lat'] * 1000) / 1000
drop_10["start_lng"] = np.floor(drop_10["start_lng"] * 1000) / 1000
drop_10["end_lat"] = np.floor(drop_10["end_lat"] * 1000) / 1000
drop_10["end_lng"] = np.floor(drop_10["end_lng"] * 1000) / 1000
```
